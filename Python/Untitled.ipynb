```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# File path
file_path = r"C:\Users\Lenovo\Desktop\All\2_AUClasses\Thesis\Python\DI.xlsx"

# Read Excel file
df = pd.read_excel(file_path, sheet_name="Sheet1")

# Rename relevant columns
df = df.rename(columns={
    'Name2': 'Participant_Name',
    'Vision correction (e.g., glasses/contact lenses)': 'Vision_Correction',
    'Experience with VR/AR': 'VR_AR_Experience',
    'Experience with Gaze in VR': 'Gaze_VR_Experience'
})

# Count and calculate percentages
gender_counts = df['Gender'].value_counts(dropna=False)
gender_percent = round(gender_counts / gender_counts.sum() * 100, 1)

# Bar plot with percentages
plt.figure(figsize=(6, 5))
sns.barplot(x=gender_percent.index, y=gender_percent.values, palette="pastel")
plt.ylabel("Percentage")
plt.title("Gender Distribution (%)")

# Add percentage labels
for i, v in enumerate(gender_percent.values):
    plt.text(i, v + 1, f"{v}%", ha='center', fontweight='bold')

plt.ylim(0, 100)
plt.tight_layout()
plt.show()
# Pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_percent, labels=gender_percent.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Gender Distribution (%)")
plt.show()
```

```python
# NLTK Sentiment Analysis for User Feedback
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import numpy as np
import re

# Download required NLTK data (run this once)
nltk.download('vader_lexicon')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

# Initialize sentiment analyzer and other tools
sia = SentimentIntensityAnalyzer()
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    """Clean and preprocess text"""
    if pd.isna(text) or text == '':
        return ''
    
    # Convert to lowercase and remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', str(text).lower())
    
    # Tokenize and remove stopwords
    tokens = word_tokenize(text)
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words and len(token) > 2]
    
    return ' '.join(tokens)

def analyze_sentiment(text):
    """Analyze sentiment using VADER"""
    if pd.isna(text) or text == '':
        return {'compound': 0, 'pos': 0, 'neu': 1, 'neg': 0}
    
    scores = sia.polarity_scores(str(text))
    return scores

def categorize_sentiment(compound_score):
    """Categorize sentiment based on compound score"""
    if compound_score >= 0.05:
        return 'Positive'
    elif compound_score <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

# Load your Excel file
df = pd.read_excel(r'C:\Users\Lenovo\Desktop\All\2_AUClasses\Thesis\Python\App.xlsx')

print(f" Loaded {len(df)} rows of data")
print(f" Columns: {list(df.columns)}")
print(f" First few rows:")
print(df.head())

# Identify the correct column names
print(f"\n Identifying column names...")

# Common variations for ID column
id_column = None
for col in df.columns:
    if any(keyword in col.lower() for keyword in ['id', 'number', 'index']):
        id_column = col
        break

# Common variations for feedback column
feedback_column = None
for col in df.columns:
    if any(keyword in col.lower() for keyword in ['feedback', 'comment', 'response', 'review']):
        feedback_column = col
        break

# If we can't find them automatically, let's use the first and last columns as defaults
if id_column is None:
    id_column = df.columns[0]  # Usually the first column is ID
    print(f" Using '{id_column}' as ID column")

if feedback_column is None:
    feedback_column = df.columns[-1]  # Usually the last column is feedback
    print(f" Using '{feedback_column}' as feedback column")

print(f" ID Column: '{id_column}'")
print(f" Feedback Column: '{feedback_column}'")

# Clean the feedback column
df['cleaned_feedback'] = df[feedback_column].apply(clean_text)

# Perform sentiment analysis
sentiment_scores = df[feedback_column].apply(analyze_sentiment)

# Extract individual scores
df['compound_score'] = [score['compound'] for score in sentiment_scores]
df['positive_score'] = [score['pos'] for score in sentiment_scores]
df['neutral_score'] = [score['neu'] for score in sentiment_scores]
df['negative_score'] = [score['neg'] for score in sentiment_scores]

# Categorize sentiments
df['sentiment_category'] = df['compound_score'].apply(categorize_sentiment)

# Display results
print("=== SENTIMENT ANALYSIS RESULTS ===\n")

print("Sample of analyzed feedback:")
print(df[[id_column, feedback_column, 'compound_score', 'sentiment_category']].head(10))

print(f"\n=== OVERALL SENTIMENT DISTRIBUTION ===")
sentiment_counts = df['sentiment_category'].value_counts()
print(sentiment_counts)

print(f"\n=== DETAILED STATISTICS ===")
print(f"Average Compound Score: {df['compound_score'].mean():.3f}")
print(f"Average Positive Score: {df['positive_score'].mean():.3f}")
print(f"Average Neutral Score: {df['neutral_score'].mean():.3f}")
print(f"Average Negative Score: {df['negative_score'].mean():.3f}")

# Create visualizations
plt.figure(figsize=(20, 15))

# 1. Sentiment Distribution Pie Chart
plt.subplot(3, 3, 1)
colors = ['#2ecc71', '#e74c3c', '#95a5a6']
sentiment_counts.plot(kind='pie', autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Sentiment Distribution')
plt.ylabel('')

# 2. Compound Scores Distribution
plt.subplot(3, 3, 2)
plt.hist(df['compound_score'], bins=15, alpha=0.7, color='skyblue', edgecolor='black')
plt.axvline(df['compound_score'].mean(), color='red', linestyle='--', label='Mean')
plt.xlabel('Compound Score')
plt.ylabel('Frequency')
plt.title('Distribution of Compound Scores')
plt.legend()

# 7. Boxplot of Sentiment Scores by Category
plt.subplot(3, 3, 7)
# Prepare data for boxplot
sentiment_data = []
sentiment_labels = []
for category in ['Positive', 'Negative', 'Neutral']:
    if category in df['sentiment_category'].values:
        category_scores = df[df['sentiment_category'] == category]['compound_score']
        sentiment_data.append(category_scores)
        sentiment_labels.append(category)

# Create boxplot
box_plot = plt.boxplot(sentiment_data, labels=sentiment_labels, patch_artist=True)
colors = ['#2ecc71', '#e74c3c', '#95a5a6']
for patch, color in zip(box_plot['boxes'], colors[:len(sentiment_labels)]):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)

plt.ylabel('Compound Score')
plt.title('Boxplot: Sentiment Scores by Category')
plt.grid(True, alpha=0.3)

# 8. Violin Plot of Sentiment Scores by Category
plt.subplot(3, 3, 8)
# Create violin plot using seaborn
if len(sentiment_labels) > 0:
    # Prepare data for violin plot
    plot_data = []
    for category in sentiment_labels:
        category_scores = df[df['sentiment_category'] == category]['compound_score']
        for score in category_scores:
            plot_data.append({'Category': category, 'Score': score})
    
    plot_df = pd.DataFrame(plot_data)
    
    # Create violin plot
    violin_parts = plt.violinplot([df[df['sentiment_category'] == cat]['compound_score'] 
                                  for cat in sentiment_labels], 
                                 positions=range(len(sentiment_labels)))
    
    # Color the violin plots
    for i, (pc, color) in enumerate(zip(violin_parts['bodies'], colors[:len(sentiment_labels)])):
        pc.set_facecolor(color)
        pc.set_alpha(0.7)
    
    plt.xticks(range(len(sentiment_labels)), sentiment_labels)
    plt.ylabel('Compound Score')
    plt.title('Violin Plot: Distribution by Category')
    plt.grid(True, alpha=0.3)

# 9. Combined Score Distribution Comparison
plt.subplot(3, 3, 9)
# Create a more detailed comparison of all sentiment components
score_types = ['Positive', 'Neutral', 'Negative', 'Compound']
score_columns = ['positive_score', 'neutral_score', 'negative_score', 'compound_score']

# Create box plots for all score types
box_data = [df[col] for col in score_columns]
box_plot2 = plt.boxplot(box_data, labels=score_types, patch_artist=True)

# Color the boxes
box_colors = ['green', 'gray', 'red', 'blue']
for patch, color in zip(box_plot2['boxes'], box_colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.6)

plt.ylabel('Score Value')
plt.title('All Sentiment Components Distribution')
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)

# 3. Sentiment Scores Comparison
plt.subplot(3, 3, 3)
scores_mean = [df['positive_score'].mean(), df['neutral_score'].mean(), df['negative_score'].mean()]
labels = ['Positive', 'Neutral', 'Negative']
bars = plt.bar(labels, scores_mean, color=['green', 'gray', 'red'], alpha=0.7)
plt.ylabel('Average Score')
plt.title('Average Sentiment Scores')
plt.ylim(0, 1)

# Add value labels on bars
for bar, score in zip(bars, scores_mean):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.01, 
             f'{score:.3f}', ha='center', va='bottom')

# 4. Sentiment Timeline
plt.subplot(3, 3, 4)
# Check if we have a time column
time_column = None
for col in df.columns:
    if any(keyword in col.lower() for keyword in ['time', 'date', 'timestamp']):
        time_column = col
        break

if time_column:
    try:
        df[time_column] = pd.to_datetime(df[time_column])
        df_sorted = df.sort_values(time_column)
        plt.plot(range(len(df_sorted)), df_sorted['compound_score'], marker='o', alpha=0.7)
        plt.xlabel('Response Order (by time)')
    except:
        plt.plot(range(len(df)), df['compound_score'], marker='o', alpha=0.7)
        plt.xlabel('Response Order')
else:
    plt.plot(range(len(df)), df['compound_score'], marker='o', alpha=0.7)
    plt.xlabel('Response Order')

plt.axhline(y=0, color='black', linestyle='-', alpha=0.3)
plt.axhline(y=0.05, color='green', linestyle='--', alpha=0.5, label='Positive threshold')
plt.axhline(y=-0.05, color='red', linestyle='--', alpha=0.5, label='Negative threshold')
plt.ylabel('Compound Score')
plt.title('Sentiment Over Time')
plt.legend()

# 5. Word Frequency Analysis for Positive and Negative Feedback
plt.subplot(3, 3, 5)
positive_feedback = df[df['sentiment_category'] == 'Positive']['cleaned_feedback'].str.cat(sep=' ')
negative_feedback = df[df['sentiment_category'] == 'Negative']['cleaned_feedback'].str.cat(sep=' ')

if positive_feedback:
    positive_words = positive_feedback.split()
    positive_word_freq = Counter(positive_words).most_common(10)
    words, counts = zip(*positive_word_freq) if positive_word_freq else ([], [])
    plt.barh(words, counts, color='green', alpha=0.7)
    plt.xlabel('Frequency')
    plt.title('Top Words in Positive Feedback')
    plt.gca().invert_yaxis()

# 6. Detailed Sentiment Breakdown
plt.subplot(3, 3, 6)
sentiment_details = df.groupby('sentiment_category')[['positive_score', 'neutral_score', 'negative_score']].mean()
sentiment_details.plot(kind='bar', ax=plt.gca(), color=['green', 'gray', 'red'], alpha=0.7)
plt.title('Average Score Components by Sentiment Category')
plt.xlabel('Sentiment Category')
plt.ylabel('Average Score')
plt.xticks(rotation=45)
plt.legend()

plt.tight_layout()
plt.show()

# Identify most positive and negative feedback
print(f"\n=== MOST POSITIVE FEEDBACK ===")
most_positive = df.loc[df['compound_score'].idxmax()]
print(f"Score: {most_positive['compound_score']:.3f}")
print(f"Feedback: {most_positive[feedback_column]}")

print(f"\n=== MOST NEGATIVE FEEDBACK ===")
most_negative = df.loc[df['compound_score'].idxmin()]
print(f"Score: {most_negative['compound_score']:.3f}")
print(f"Feedback: {most_negative[feedback_column]}")

# Key insights extraction
print(f"\n=== KEY INSIGHTS ===")
total_responses = len(df)
positive_pct = (sentiment_counts.get('Positive', 0) / total_responses) * 100
negative_pct = (sentiment_counts.get('Negative', 0) / total_responses) * 100
neutral_pct = (sentiment_counts.get('Neutral', 0) / total_responses) * 100

print(f" {positive_pct:.1f}% of feedback is positive")
print(f" {negative_pct:.1f}% of feedback is negative") 
print(f" {neutral_pct:.1f}% of feedback is neutral")
print(f" Overall sentiment is {'positive' if df['compound_score'].mean() > 0.05 else 'negative' if df['compound_score'].mean() < -0.05 else 'neutral'}")
print(f" Average sentiment strength: {abs(df['compound_score'].mean()):.3f}")

# Export results to CSV
df_export = df[[id_column, feedback_column, 'compound_score', 'positive_score', 
                'neutral_score', 'negative_score', 'sentiment_category']]
df_export.to_csv('sentiment_analysis_results.csv', index=False)
print(f"\n Results exported to 'sentiment_analysis_results.csv'")

# Additional analysis: Extract key themes
print(f"\n=== THEME ANALYSIS ===")
all_feedback = ' '.join(df['cleaned_feedback'].dropna())
if all_feedback:
    word_freq = Counter(all_feedback.split()).most_common(15)
    print("Most common words in feedback:")
    for word, count in word_freq:
        print(f"  {word}: {count} times")
```
